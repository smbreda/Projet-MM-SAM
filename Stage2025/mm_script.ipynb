```python
import os
import cv2
import random
import torch
from PIL import Image
```

```python
# Dossiers
shared_dir = "/content/drive/MyDrive/Stage2025/shared"
os.makedirs(shared_dir, exist_ok=True)
os.makedirs(os.path.join(shared_dir, "mmpose"), exist_ok=True)
```

```python
# Extraction aléatoire d'une frame depuis une vidéo
def extract_random_frame(video_path, output_path):
    try:
        video = cv2.VideoCapture(video_path)
        total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
        if total_frames <= 0:
            print("Error: Could not retrieve frame count from the video.")
            return

        random_frame_number = random.randint(0, total_frames - 1)
        video.set(cv2.CAP_PROP_POS_FRAMES, random_frame_number)

        ret, frame = video.read()
        if ret:
            cv2.imwrite(output_path, frame)
            print(f"Random frame saved as {output_path}")
        else:
            print("Error: Could not read the frame.")

        video.release()
    except Exception as e:
        print(f"An error occurred: {e}")

# Chemins
video_file = "/content/testAlly.mp4"
image_path = os.path.join(shared_dir, "mmpose", "random_frame.jpg")

# Extraction frame
extract_random_frame(video_file, image_path)
```

```python
from mmpose.apis import inference_topdown, init_model
from segment_anything import sam_model_registry, SamPredictor
```

```python
# Initialisation MMPose
config_path = "configs/body_2d_keypoint/topdown_heatmap/coco/td-hm_hrnet-w48_8xb32-210e_coco-256x192.py"
checkpoint_path = "td-hm_hrnet-w48_8xb32-210e_coco-256x192-0e67c616_20220913.pth"
model = init_model(config_path, checkpoint_path, device="cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Inference
results = inference_topdown(model, image_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extraction du barycentre
keypoints = results[0].pred_instances.keypoints[0]
xy_points = keypoints[:, :2]
barycenter = np.mean(xy_points, axis=0)
input_point = np.array([barycenter])
input_label = np.array([1])
```

```python
# Segment Anything
sam_checkpoint = "/content/sam_vit_h_4b8939.pth"
model_type = "vit_h"
device = "cuda" if torch.cuda.is_available() else "cpu"

sam = sam_model_registry[model_type](checkpoint=sam_checkpoint)
sam.to(device=device)
predictor = SamPredictor(sam)
```

```python
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
predictor.set_image(image_rgb)
```

```python
masks, scores, logits = predictor.predict(
    point_coords=input_point,
    point_labels=input_label,
    multimask_output=True
)

final_mask = masks[2]  # Choix du masque le plus probable
```

```python
# Sauvegarde du masque
mask_path = os.path.join(shared_dir, "mmpose", "mask.png")
Image.fromarray((final_mask * 255).astype(np.uint8)).save(mask_path)

print("Masque et image sauvegardés dans le dossier partagé.")
```
